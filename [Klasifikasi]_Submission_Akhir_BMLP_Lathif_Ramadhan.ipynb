# **3. Data Splitting**

# **4. Membangun Model Klasifikasi**


## **a. Membangun Model Klasifikasi**

Setelah memilih algoritma klasifikasi yang sesuai, langkah selanjutnya adalah melatih model menggunakan data latih.

Berikut adalah rekomendasi tahapannya.
1. Pilih algoritma klasifikasi yang sesuai, seperti Logistic Regression, Decision Tree, Random Forest, atau K-Nearest Neighbors (KNN).
2. Latih model menggunakan data latih.

```python
# Inisialisasi model Random Forest
model = RandomForestClassifier(random_state=42)

# Latih model menggunakan data latih
model.fit(X_train, y_train)
```

Tulis narasi atau penjelasan algoritma yang Anda gunakan.

Kode yang saya gunakan memanfaatkan algoritma **Random Forest Classifier** untuk melakukan klasifikasi. Sederhananya, bayangkan kita memiliki data pelanggan dan ingin mengelompokkannya ke dalam *cluster* berdasarkan karakteristik mereka. Random Forest akan membantu dalam tugas ini.

**Cara Kerja:**

1. **Kekuatan Bersama (Ensemble Learning)**: Random Forest termasuk dalam keluarga algoritma *Ensemble Learning*. Artinya, ia menggabungkan prediksi dari banyak model sederhana (**Decision Tree/Pohon Keputusan**) untuk menghasilkan prediksi yang lebih akurat dan andal. Ibaratnya, kita meminta pendapat dari banyak ahli, bukan hanya satu, untuk mendapatkan keputusan yang lebih baik.

2. **Membangun Pohon Keputusan:**

  * Random Forest menciptakan banyak pohon keputusan.
  * Setiap pohon dilatih dengan data yang berbeda, dipilih secara acak dari data asli (**bootstrap sampling**). Seperti membagi data menjadi beberapa bagian dan memberikannya ke pohon-pohon berbeda untuk dipelajari.
  * Saat membangun setiap pohon, hanya beberapa fitur yang dipilih secara acak untuk dipertimbangkan di setiap percabangan (**feature randomness**). Ini membuat pohon-pohon memiliki karakteristik yang berbeda dan mengurangi kemungkinan salah prediksi karena terlalu fokus pada data latihan.

**3. Membuat Prediksi:**

  * Ketika ada data baru yang ingin diprediksi, data tersebut akan diproses oleh setiap pohon keputusan di dalam *forest*.
  * Setiap pohon memberikan prediksinya sendiri.
  * Untuk tugas klasifikasi (seperti dalam kode ini), prediksi akhir ditentukan melalui **voting**. Kelas/cluster yang mendapatkan suara terbanyak dari pohon-pohon tersebut akan menjadi prediksi akhirnya.


## **b. Evaluasi Model Klasifikasi**

Berikut adalah **rekomendasi** tahapannya.
1. Lakukan prediksi menggunakan data uji.
2. Hitung metrik evaluasi seperti Accuracy dan F1-Score (Opsional: Precision dan Recall).
3. Buat confusion matrix untuk melihat detail prediksi benar dan salah.

```python
# Lakukan prediksi pada data uji
y_pred = model.predict(X_test)
```

```python
# Hitung metrik evaluasi
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted') # Gunakan average='weighted' untuk multi-class
```

```python
# Cetak hasil evaluasi
print("Accuracy:", accuracy)
print("F1-Score:", f1)
```

**Arti Accuracy dan F1-Score 1.0**

* **Accuracy 1.0**: Artinya model saya berhasil memprediksi semua data di data uji dengan benar. Tidak ada data yang salah diklasifikasikan.
* **F1-Score 1.0**: F1-Score adalah metrik yang menggabungkan precision dan recall.   Nilai 1.0 menunjukkan bahwa model saya memiliki precision dan recall yang sempurna, yang berarti:
  * **Precision 1.0**: Semua data yang diprediksi sebagai positif (milik kelas tertentu) memang benar-benar positif. Tidak ada False Positive.
  * **Recall 1.0**: Semua data yang sebenarnya positif berhasil diidentifikasi oleh model. Tidak ada False Negative.

Hasil ini mengindikasikan bahwa model saya memiliki performa yang sangat baik pada data uji. Model mampu memprediksi semua kelas dengan sempurna.

```python
# Buat dan tampilkan confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

**Penjelasan Elemen Matriks**

* **[125, 0, 0]**:
  * 125 data yang sebenarnya milik kelas 1 diprediksi dengan benar sebagai kelas 1.
  * 0 data yang sebenarnya milik kelas 1 diprediksi salah sebagai kelas 2.
  * 0 data yang sebenarnya milik kelas 1 diprediksi salah sebagai kelas 3.
* **[0, 156, 0]**:
  * 0 data yang sebenarnya milik kelas 2 diprediksi salah sebagai kelas 1.
  * 156 data yang sebenarnya milik kelas 2 diprediksi dengan benar sebagai kelas 2.
  * 0 data yang sebenarnya milik kelas 2 diprediksi salah sebagai kelas 3.
* [0, 0, 167]:
  * -0 data yang sebenarnya milik kelas 3 diprediksi salah sebagai kelas 1.
  * 0 data yang sebenarnya milik kelas 3 diprediksi salah sebagai kelas 2.
  * 167 data yang sebenarnya milik kelas 3 diprediksi dengan benar sebagai kelas 3.

**Kesimpulan**

Dari Confusion Matrix ini, dapat disimpulkan bahwa model saya memiliki performa yang **sangat baik** dalam mengklasifikasikan data.

* **Tidak ada False Positive atau False Negative**: Semua data diprediksi dengan benar ke dalam kelas yang seharusnya.
* **Diagonal Utama**: Nilai-nilai pada diagonal utama (125, 156, 167) menunjukkan jumlah prediksi yang benar untuk setiap kelas. Nilai-nilai ini tinggi, yang menunjukkan performa model yang baik.
* **Nilai di Luar Diagonal**: Nilai-nilai di luar diagonal utama adalah 0, yang berarti tidak ada kesalahan klasifikasi.

Hasil ini konsisten dengan Accuracy dan F1-Score 1.0 yang saya dapatkan sebelumnya. Model saya mampu membedakan dan mengklasifikasikan ketiga kelas dengan sempurna pada data uji.

```python
# Tampilkan classification report untuk detail lebih lanjut
print(classification_report(y_test, y_pred))
```

**Istilah dalam Classification Report**

* **Precision**: Proporsi data yang diprediksi positif dan benar-benar positif.
  * Rumus: `Precision = True Positive / (True Positive + False Positive)`
* **Recall**: Proporsi data positif yang berhasil diidentifikasi oleh model.
  * Rumus: `Recall = True Positive / (True Positive + False Negative)`
* F1-Score: Rata-rata harmonik dari precision dan recall. Memberikan keseimbangan antara keduanya.
  * Rumus: `F1-Score = 2 * (Precision * Recall) / (Precision + Recall)`
* **Support**: Jumlah data aktual untuk setiap kelas.

**Interpretasi Classification Report**

Report ini menunjukkan performa model untuk setiap kelas (`0`, `1`, dan `2`) dan secara keseluruhan.

* **Kelas 0**:

  * Precision: 1.00 (Semua data yang diprediksi sebagai kelas 0 memang benar-benar kelas 0)
  * Recall: 1.00 (Semua data yang sebenarnya kelas 0 berhasil diidentifikasi)
  * F1-Score: 1.00 (Performa sempurna untuk kelas 0)
  * Support: 125 (Jumlah data aktual untuk kelas 0)
* **Kelas 1 dan 2**: Pola yang sama dengan kelas 0, menunjukkan performa sempurna untuk kedua kelas ini juga.
* **Accuracy**: 1.00 (Akurasi keseluruhan model, yang berarti semua data diprediksi dengan benar)

* **macro avg**: Rata-rata precision, recall, dan F1-Score untuk semua kelas tanpa mempertimbangkan jumlah data per kelas.

* **weighted avg**: Rata-rata precision, recall, dan F1-Score untuk semua kelas dengan mempertimbangkan jumlah data per kelas.

**Kesimpulan**

Classification Report ini menegaskan bahwa model saya memiliki **performa yang sangat baik** dalam mengklasifikasikan data ke dalam tiga kelas. Semua metrik (`precision`, `recall`, `F1-Score`) bernilai **1.00** untuk semua kelas, menunjukkan tidak ada kesalahan klasifikasi pada data uji.
